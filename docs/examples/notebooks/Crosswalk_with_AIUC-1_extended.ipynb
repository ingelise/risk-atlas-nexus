## 1. Overview of AIUC-1 resources

### Hierarchy Structure

- **6 Principles** (A-F)
  - Contains 51 Requirements (Rules)
    - Each Requirement contains 1-3 Control Activities (Rules)
      - Total: 129 Control Activities

### Class Details

#### Core Classes

| Class           | Purpose                                                             | Count |
| --------------- | ------------------------------------------------------------------- | ----- |
| Principle       | Top-level AIUC-1 framework principle                                | 6     |
| Requirement     | Specific requirement associated with a principle                    | 51    |
| ControlActivity | Concrete control (Permission/Prohibition/Obligation/Recommendation) | 129   |

## 3. AIUC-1 × IBM AI Risk Atlas
Let's automatically create a crosswalk from the mappings provided.

### 3.1 IBM AI Risk Atlas crosswalks by risk

```python
# Util functions

def get_mapped_aiuc1_requirements(ibm_risk_id, aiuc1_requirements):
    """
    Find the AIUC-1 requirements that map to a given IBM risk
    """
    mapping_fields = ['close_mappings', 'exact_mappings', 'related_mappings', 'broad_mappings', 'narrow_mappings']
    matched_reqs = defaultdict(list)  # req_id -> [mapping_types]
    
    for req in aiuc1_requirements:
        for field_name in mapping_fields:
            if hasattr(req, field_name):
                field_value = getattr(req, field_name)
                if field_value:
                    ids = field_value if isinstance(field_value, list) else [field_value]
                    if ibm_risk_id in ids:
                        if req.id not in matched_reqs:
                            matched_reqs[req.id] = []
                        matched_reqs[req.id].append(field_name.replace('_mappings', '').capitalize())
    
    return matched_reqs

# Get all AIUC-1 requirements
aiuc1_requirements = ran.get_all(class_name='requirement', taxonomy='aiuc1')

# Get all IBM Risk Atlas risks
ibm_risks = ran.get_all(class_name='entries', taxonomy='ibm-risk-atlas')
ibm_risks = [r for r in ibm_risks if hasattr(r, 'isDefinedByTaxonomy') and r.isDefinedByTaxonomy == 'ibm-risk-atlas']

# Create lookup dictionaries
ibm_risk_map = {risk.id: risk for risk in ibm_risks}
aiuc1_req_map = {req.id: req for req in aiuc1_requirements}


# Build the crosswalk data 
crosswalk_data = []

for ibm_risk in ibm_risks:
    mapped_aiuc1_reqs = get_mapped_aiuc1_requirements(ibm_risk.id, aiuc1_requirements)
    mapped_reqs_list = []
    for req_id, mapping_types in sorted(mapped_aiuc1_reqs.items()):
        req = aiuc1_req_map.get(req_id)
        if req:
            mapping_str = ', '.join(mapping_types)
            mapped_reqs_list.append(f'{req_id} ({mapping_str})')
    
    crosswalk_data.append({
        'IBM_AI_Risk_Atlas_Risk_ID': ibm_risk.id,
        'IBM_AI_Risk_Atlas_Risk_Name': ibm_risk.name,
        'IBM_Description': ibm_risk.description[:150] if ibm_risk.description else '',
        'Mapped_AIUC1_Requirements_Count': len(mapped_aiuc1_reqs),
        'Mapped_AIUC1_Requirements': ', '.join(mapped_reqs_list) if mapped_reqs_list else ''
    })

crosswalk_df = pd.DataFrame(crosswalk_data)

# Sort by number of mappings (descending)
crosswalk_df = crosswalk_df.sort_values('Mapped_AIUC1_Requirements_Count', ascending=False).reset_index(drop=True)
print("Crosswalk ready")
```

## 4. Use Case to Risks to Controls Mapping

This section demonstrates how to take a use case description, identify associated risks from the IBM AI Risk Atlas, and then map those risks to AIUC-1 controls.

```python
from ai_atlas_nexus.blocks.inference import (

    OllamaInferenceEngine,

)
from ai_atlas_nexus.blocks.inference.params import (
    InferenceEngineCredentials,
 
    OllamaInferenceEngineParams,

)


from typing import List, Dict
import re

def extract_risks_from_usecase(usecase_string: str, ran_instance, inference_engine=None) -> List[Dict]:
    """
    Identify IBM AI Risk Atlas risks 
    
    Args:
        usecase_string: Description of an AI use case
        ran_instance: AIAtlasNexus instance
        inference_engine: inference engine 
    
    Returns:
        List of dictionaries with risk information
    """
    matched_risks = []
    
    # If inference engine is provided, use LLM-based identification
    if inference_engine is not None:
        try:
            risks = ran_instance.identify_risks_from_usecases(
                usecases=[usecase_string],
                inference_engine=inference_engine,
                taxonomy="ibm-risk-atlas",
                max_risk=10,
                zero_shot_only=False
            )
            
            if risks and len(risks) > 0:
                for risk in risks[0]:
                    matched_risks.append({
                        'risk_id': risk.id,
                        'risk_name': risk.name,
                        'risk_description': risk.description[:200] if risk.description else '',
                      
                    
                    })
            
            return matched_risks
        except Exception as e:
            print(f" LLM-based identification failed ({str(e)[:50]}...).")


def get_controls_for_risks(risk_ids: List[str], aiuc1_requirements: List) -> Dict[str, List[Dict]]:
    """
    Get AIUC-1 controls (rules) for a list of IBM risks.
    
    Args:
        risk_ids: List of IBM risk IDs
        aiuc1_requirements: List of AIUC-1 requirement objects
    
    Returns:
        Dictionary mapping risk_id -> list of controls
    """
    risk_to_controls = {}
    #TODO compact this
    for risk_id in risk_ids:
        controls = []
        for req in aiuc1_requirements:
            # Check if risk is in any of the mapping fields
            mapping_fields = ['close_mappings', 'exact_mappings', 'related_mappings', 'broad_mappings', 'narrow_mappings']
            for field_name in mapping_fields:
                if hasattr(req, field_name):
                    field_value = getattr(req, field_name)
                    if field_value and risk_id in (field_value if isinstance(field_value, list) else [field_value]):
                        # Get the control rules for this requirement
                        if hasattr(req, 'hasRule') and req.hasRule:
                            rules = req.hasRule if isinstance(req.hasRule, list) else [req.hasRule]
                            for rule_id in rules:
                                rule = ran.get_by_id(class_name="rules", identifier=rule_id)
                                if rule:
                                    controls.append({
                                        'requirement_id': req.id,
                                        'requirement_name': req.name,
                                        'control_id': rule.id,
                                        'control_name': rule.name if hasattr(rule, 'name') else rule_id,
                                        'control_description': rule.description if hasattr(rule, 'description') else '',
                                        'mapping_type': field_name.replace('_mappings', '')
                                    })
        
        risk_to_controls[risk_id] = controls
    
    return risk_to_controls


def map_usecase_to_controls(usecase_string: str, ran_instance, inference_engine=None, top_k: int = 10) -> pd.DataFrame:
    """
    Map Use case -> IBM Risks -> AIUC-1 Controls
    
    Args:
        usecase_string: Description of an AI use case
        ran_instance: AIAtlasNexus instance
        inference_engine: inference engine
        top_k: Maximum number of top risks to return
    
    Returns:
        DataFrame with use case, risks, and associated AIUC-1 controls
    """

    aiuc1_requirements = ran_instance.get_all(class_name='requirement', taxonomy='aiuc1')
    matched_risks = extract_risks_from_usecase(usecase_string, ran_instance, inference_engine)
    matched_risks = matched_risks[:top_k]  # Keep top K
    
    if not matched_risks:
        return pd.DataFrame()
    
    risk_ids = [r['risk_id'] for r in matched_risks]
    risk_to_controls = get_controls_for_risks(risk_ids, aiuc1_requirements)
    
    results = []
    for risk_info in matched_risks:
        risk_id = risk_info['risk_id']
        controls = risk_to_controls.get(risk_id, [])
        
        if controls:
            for control in controls:
                results.append({
                    'Use_Case': usecase_string[:100],
                    'IBM_Risk_ID': risk_id,
                    'IBM_Risk_Name': risk_info['risk_name'],
                    'Requirement_Mapping_Type': control['mapping_type'],
                    'AIUC1_Requirement_ID': control['requirement_id'],
                    'AIUC1_Requirement_Name': control['requirement_name'],
                    'AIUC1_Control_ID': control['control_id'],
                    'AIUC1_Control_Name': control['control_name'],
                    
                })
        else:
            # Risk with no controls mapped
            results.append({
                'Use_Case': usecase_string[:100],
                'IBM_Risk_ID': risk_id,
                'IBM_Risk_Name': risk_info['risk_name'],
                'Requirement_Mapping_Type': None,
                'AIUC1_Requirement_ID': None,
                'AIUC1_Requirement_Name': None,
                'AIUC1_Control_ID': None,
                'AIUC1_Control_Name': None,
                
            })
    
    return pd.DataFrame(results)


print("Use case mapping functions loaded")
```

### 4.1 Example: Financial Chatbot Use Case

Let's demonstrate the use case mapping with a financial chatbot example.

```python
# Example 1: Financial chatbot use case
usecase_1 = """
A financial services company is building a generative AI chatbot to provide 
investment advice to customers. The system uses large language models trained on 
financial data. Key concerns include: preventing hallucinations about market data, 
ensuring data privacy for customer information, securing against prompt injection 
attacks from adversaries, and preventing unauthorized access to sensitive financial records.
The system must maintain audit trails and ensure reliability of tool calls to external systems.
"""


# Option B: Using LLM-based identification (requires inference engine)
inference_engine = OllamaInferenceEngine(
    model_name_or_path="granite3.3:8b",
    credentials=InferenceEngineCredentials(api_url="localhost:11434"),
    parameters=OllamaInferenceEngineParams(
        num_predict=1000, num_ctx=8192, temperature=0, repeat_penalty=1
    )
)


result_df_1 = map_usecase_to_controls(usecase_1, ran, inference_engine=inference_engine)

print(f"\n✓ Identified {result_df_1['IBM_Risk_ID'].nunique()} risks and {len(result_df_1)} control mappings")
print(f"\nTop risks identified:")
print(result_df_1[['IBM_Risk_ID', 'IBM_Risk_Name']].drop_duplicates('IBM_Risk_ID').head(10).to_string(index=False))

print(f"\n\nFull mapping results:")
result_df_1
```

### 4.4 Analyzing Results

```python
def summarize_usecase_analysis(result_df: pd.DataFrame) -> Dict:
    """
    Generate a summary 
    
    Args:
        result_df: DataFrame returned by map_usecase_to_controls()
    
    Returns:
        Dictionary with summary statistics
    """
    if result_df.empty:
        return {"status": "No risks identified"}
    
    summary = {
        "total_risks_identified": result_df['IBM_Risk_ID'].nunique(),
        "total_controls_mapped": len(result_df),
        "risks_with_no_controls": (result_df['AIUC1_Control_ID'].isna()).sum(),
        "risks_with_controls": (result_df['AIUC1_Control_ID'].notna()).sum(),
        "unique_requirements": result_df['AIUC1_Requirement_ID'].nunique(),
        "mapping_types": result_df['Mapping_Type'].value_counts().to_dict() if 'Mapping_Type' in result_df.columns else {},
        "top_5_risks": result_df[['IBM_Risk_ID', 'IBM_Risk_Name']].drop_duplicates('IBM_Risk_ID').head(5).values.tolist()
    }
    return summary


# Analyze financial chatbot results

print("FINANCIAL CHATBOT USE CASE - SUMMARY")

summary_1 = summarize_usecase_analysis(result_df_1)
for key, value in summary_1.items():
    if key != 'top_5_risks':
        print(f"{key}: {value}")
print("\nTop 5 Identified Risks:")
for i, (risk_id, risk_name) in enumerate(summary_1['top_5_risks'], 1):
    print(f"  {i}. {risk_id}: {risk_name}")


print("HR RECRUITMENT SYSTEM USE CASE - SUMMARY")

summary_2 = summarize_usecase_analysis(result_df_2)
for key, value in summary_2.items():
    if key != 'top_5_risks':
        print(f"{key}: {value}")
print("\nTop 5 Identified Risks:")
for i, (risk_id, risk_name) in enumerate(summary_2['top_5_risks'], 1):
    print(f"  {i}. {risk_id}: {risk_name}")
```
